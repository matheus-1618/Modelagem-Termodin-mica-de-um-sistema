# Validação do Modelo#



O objetivo deste experimento é basicamente implementar um modelo para um sistema termodinâmico constituído por um recepiente, inicialmente de cerâmica, porcelana AAA que ao ser retirado de temperatura 0ºC, terá de retornar a temperatura do ambiente de experimentação, na ocasião, por volta de 23,1ºC. Destarte, foram estabelecidas hipóteses com entuito de diminuir o tempo para esse retorno, como a mudança das dimensões do recepiente, mantido o volume constante, utilização de recepientes de materiais diferentes e até outros líquidos, com diferentes propriedades físicas.

As possíveis perguntas para a modelação então orbitaram em torno de como atingir o "combo" recepiente, material e líquido que pudessem minimizar o tempo de equilíbrio térmico entre líquido e ambiente externo.

Para o experimento, utilizaremos um recipiente de altura $9cm$, mas $3cm$ de líquido ocupado, para manter $150$ml de liquido como constante, com raio $4cm$, com espessura de $5mm$.

Nossas equações foram implementadas da seguinte maneira:

$\frac{dT_Liquido}{dt}=\frac{1}{m_o\cdot c_o}\cdot[\dot{Q_2}+\dot{Q_3}]$

$\frac{dT_Copo}{dt}=\frac{1}{m_r\cdot c_r}\cdot[\dot{Q_1}-\dot{Q_2}]$

Sendo:

$\dot{Q_1}=\frac{T_a-T_r}{\frac{e}{k_c\cdot A_r}+\frac{1}{h_c\cdot A_c}}$

$\dot{Q_2}=\frac{T_r-T_o}{\frac{e}{k_c\cdot A_c}}$

$\dot{Q_3}=\frac{T_a-T_o}{\frac{1}{h_a\cdot A_s}}$




Abaixo, inicializamos todos os parâmetros do problema. Estes poderão ser alterados mais abaixo, com entuito máximo de conseguir observar variações interessantes às perguntas estabelecidas no começo. Todos os parâmetros estão em unidades coerentes que nos permitem trabalhar sem quaisquer erro de unidades no sistema, gerando resultados desproporcionais.

```python
from math import pi
#densidade da agua - 0,997 kg/L --> 997 kg/m3
do=997
#densidade do copo (cerâmica/porcelana)
drec=2300
#volume de água - fixado em 150 cm³, com altura de 3cm
vo=150*(0.000001)
#volume do recipiente-Considerando nesse ponto, o analisado
vr=150*(0.000001)
#massa de água
mo=do*vo
#massa do recipiente
mr=drec*vr
#calor específico da água
co=4190
#calor específico da cerâmica
cr=1085
#temperatura ambiente, em kelvin
Ta=23.1+273.15
#temperatura inicial do experimento, em kelvin
Ti=0+273.15
#espessura do recipiente: 5mm --> 0.005m: vamos dividi-lo ao meio na modelagem da condução térmica
e=0.005/2
#área da superfície de contato entre a água e o ar(pi.r²)
As=pi*(0.04)**2
#área de contato das paredes externas do recipiente(2.pi.r.h)
Ar=2*pi*(0.04)*(0.03)
#área de contato das paredes internas recipiente (vamos simplificar e igualá-las aqui)
Ac=Ar
#condutividade térmica da cerâmica/porcelana --> 3W/mk
kc=3
#coeficiente de transferência convectiva da superfície do ar para o água --> 5-30W/m2k
ha=15.5
#coeficiente de transferência convectiva do ar para o recipiente --> 5-30W/m2k
hc=15.5
```

<left><b>Implementação da função</b></left>
------------------------------------------------------------------------------------------------------------------------------

Utilizando as equações estabelecidas acima, pela transmissão de calor, chegamos a implementação da função do modelo que auxiliará na construção do modelo.

```python
def modelo_caneca(temperatura,tempo):#função do nosso modelo
    To=temperatura[0]#temperatura do líquido
    Tr=temperatura[1]#temperatura do recepiente
    Q1=(Ta-Tr)/((e/(kc*Ar))+(1/(hc*Ar)))#equação do fluxo 1
    Q2=(Tr-To)/(e/(kc*Ac))#equação do fluxo 2
    Q3=(Ta-To)/(1/(ha*As))#equação do fluxo 3
    
    dTodt=(Q2+Q3)/(mo*co)#variação instântanea da temperatura do líquido
    dTrdt=(Q1-Q2)/(mr*cr)#variação instântanea da temperatura do recepiente
    dxdt=[dTodt,dTrdt]
    
    return dxdt
```

<left><b>Iterando o Sistema</b></left>
------------------------------------------------------------------------------------------------------------------------------

Utilizando a função odeint, e as mesmas faixas temporais utilizados acima, compararemos nosso modelo para ver se este realmente se faz coerente com os dados obtidos.


```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

To=Ti
Tr=Ti
listinicial=[To,Tr]#Definindo os parâmetros de temperatura iniciais (0ºC ou 273.15k)

tempo1=np.arange(0,22800,120)#lista de tempo utilziada, de minutos para segundos
mododeint=odeint(modelo_caneca,listinicial,tempo1)#função odeint utilizada
a1=[x-273.15 for x in mododeint[:,0]]#conversão de celsius para Kelvin
b1=[x-273.15 for x in mododeint[:,1]]#conversão de celsius para Kelvin
c1=[n/60 for n in tempo1]#conversão de segundos para minutos

ax=plt.axes()
plt.plot(dados_equilibrio2()[0],(dados_equilibrio2()[1]),'r',label='To Medida')
plt.plot(c1,a1,'k',label='To calculada')
plt.plot(c1,b1,'y',label='Tr Calculada')

plt.legend(loc="best")
ax.set_facecolor('xkcd:white')
plt.title("Modelo versus medição de 2 minutos")
plt.xlabel('Tempo(min)')
plt.ylabel('Temperatura-Celsius')
plt.grid(True)
plt.show()
```

<left><b>Modificando dimensões</b></left>
------------------------------------------------------------------------------------------------------------------------------


```python
#Alterando o modelo para variação dos parâmetros dimensionais de raio e altura, com a condição de manter o volume constante.
def modelo_caneca1(temperatura,tempo,r,h):
    To=temperatura[0]
    Tr=temperatura[1]
    #fluxos em função da altura e raio
    Q1=(Ta-Tr)/((e/(kc*(2*pi*(r)*(h)))+(1/(hc*(2*pi*(r)*(h))))))
    Q2=(Tr-To)/(e/(kc*(2*pi*(r)*(h))))
    Q3=(Ta-To)/(1/(ha*(pi*(r)**2)))
    
    dTodt=(Q2+Q3)/(mo*co)
    dTrdt=(Q1-Q2)/(mr*cr)
    dxdt=[dTodt,dTrdt]
    
    return dxdt

tempo1=np.arange(0,22800,120)                
x0=[Ta,Ta] 
#Criando dimensões que mantenham o volume de líquido utilizado constante
ri=np.arange(0.01,0.15,0.015)
hi=vo/(pi*(ri**2))
i=0
ax=plt.axes()  
while i<len(ri):
        mododeint=odeint(modelo_caneca1,listinicial,tempo1,args=(ri[i],hi[i]))
        a1=[x-273.15 for x in mododeint[:,0]]
        b1=[x-273.15 for x in mododeint[:,1]]
        c1=[n/60 for n in tempo1]
        plt.plot(c1,a1,label = 'raio= %.1f cm' %(ri[i]*100))
        i+=1

      
plt.legend(loc='best')
ax.set_facecolor('white')
plt.title('Variação de raio e altura dos recepientes')
plt.xlabel('Tempo(min)')
plt.ylabel('Temperatura(celsius)')
plt.grid(True)
plt.show()

```

```python
def modelo_caneca2(temperatura,tempo,do,co):
    To=temperatura[0]
    Tr=temperatura[1]
    Q1=(Ta-Tr)/((e/(kc*Ar))+(1/(hc*Ar)))
    Q2=(Tr-To)/(e/(kc*Ac))
    Q3=(Ta-To)/(1/(ha*As))
    #modificando a densidade e o calor específico do líquido escolhido
    dTodt=(Q2+Q3)/((do*vo)*co)
    dTrdt=(Q1-Q2)/(mr*cr)
    dxdt=[dTodt,dTrdt]
    
    return dxdt


tempo1=np.arange(0,22800,120)                
x0=[Ta,Ta] 
liqi=['café','coca','água','leite','suco','alcool','óleo']
doi=[1060,1030,997,1034,1032,789,917]
coi=[4109,4050,4190,3938.6,3494.85,2430.2,1814.9]

i=0
ax=plt.axes()  
while i<len(coi):
        mododeint=odeint(modelo_caneca2,listinicial,tempo1,args=(doi[i],coi[i]))
        a1=[x-273.15 for x in mododeint[:,0]]
        b1=[x-273.15 for x in mododeint[:,1]]
        c1=[n/60 for n in tempo1]
        plt.plot(c1,a1,label = '{0}'.format(liqi[i]))
        i+=1
        
plt.legend(loc='best')
ax.set_facecolor('lightblue')
plt.title('Variação do líquido utilizado- densidade e calor específico')
plt.xlabel('Tempo(min)')
plt.ylabel('Temperatura(celsius)')
plt.grid(True)
plt.show()
```

<left><b>Modificando o material do recepiente</b></left>

```python
def modelo_caneca3(temperatura,tempo,cr,drec,kc):
    To=temperatura[0]
    Tr=temperatura[1]
    #modificando calor específico, densidade do recepeiente e k do material
    Q1=(Ta-Tr)/((e/(kc*Ar))+(1/(hc*Ar)))
    Q2=(Tr-To)/(e/(kc*Ac))
    Q3=(Ta-To)/(1/(ha*As))
    
    dTodt=(Q2+Q3)/(mo*co)
    dTrdt=(Q1-Q2)/((drec*vr)*cr)
    dxdt=[dTodt,dTrdt]
    
    return dxdt

tempo1=np.arange(0,22800,120)                
x0=[Ta,Ta] 
cri=[1085,897,700,1250]
dreci=[2300,2700,2500,960]
kci=[3,205,0.8,0.35]
mati=['Cerâmica','Alumínio','Vidro','Polietileno']


i=0
ax=plt.axes()  
while i<len(mati):
        mododeint=odeint(modelo_caneca3,listinicial,tempo1,args=(cri[i],dreci[i],kci[i]))
        a1=[x-273.15 for x in mododeint[:,0]]
        b1=[x-273.15 for x in mododeint[:,1]]
        c1=[n/60 for n in tempo1]
        plt.plot(c1,b1,label = '{0}'.format(mati[i]))
        i+=1
        
plt.legend(loc='best')
ax.set_facecolor('lightblue')
plt.title('Variação do material utilizado- calor específico, densidade, condutividade térmica')
plt.xlabel('Tempo(min)')
plt.ylabel('Temperatura(celsius)')
plt.grid(True)
plt.show()


i=0
ax=plt.axes()  
while i<len(kci):
        mododeint=odeint(modelo_caneca3,listinicial,tempo1,args=(cri[i],dreci[i],kci[i]))
        a1=[x-273.15 for x in mododeint[:,0]]
        b1=[x-273.15 for x in mododeint[:,1]]
        c1=[n/60 for n in tempo1]
        plt.plot(c1,b1,label = '{0}'.format(mati[i]))
        i+=1


plt.legend(loc='best')
ax.set_facecolor('white')
plt.axis([125,130,17.5,19.5])
plt.title('Gráfico anterior com Zoom')
plt.xlabel('Tempo(min)')
plt.ylabel('Temperatura(celsius)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from math import pi
#densidade do óleo
do=917
#densidade do copo (polietileno)
drec=960
#volume de oleo - fixado em 150 cm³, com altura de 3cm
vo=150*(0.000001)
#volume do recipiente-Considerando nesse ponto, o analisado
vr=150*(0.000001)
#massa de oleo
mo=do*vo
#massa do recipiente
mr=drec*vr
#calor específico do óleo
co=1814
#calor específico do polietileno
cr=1250
#temperatura ambiente, em kelvin
Ta=23.1+273.15
#temperatura inicial do experimento, em kelvin
Ti=0+273.15
#espessura do recipiente: 5mm --> 0.005m: vamos dividi-lo ao meio na modelagem da condução térmica
e=0.005/2
#área da superfície de contato entre a água e o ar(pi.r²)
As=pi*(0.145)**2
#área de contato das paredes externas do recipiente(2.pi.r.h)
Ar=2*pi*(0.145)*(0.0022)
#área de contato das paredes internas recipiente (vamos simplificar e igualá-las aqui)
Ac=Ar
#condutividade térmica do polietileno 
kc=0.35
#coeficiente de transferência convectiva da superfície do ar para o óleo --> 5-30W/m2k
ha=15.5
#coeficiente de transferência convectiva do ar para o recipiente --> 5-30W/m2k
hc=15.5

def modelo_caneca4(temperatura,tempo):#função do nosso modelo-combo
    To=temperatura[0]#temperatura do líquido
    Tr=temperatura[1]#temperatura do recepiente
    Q1=(Ta-Tr)/((e/(kc*Ar))+(1/(hc*Ar)))#equação do fluxo 1
    Q2=(Tr-To)/(e/(kc*Ac))#equação do fluxo 2
    Q3=(Ta-To)/(1/(ha*As))#equação do fluxo 3
    
    dTodt=(Q2+Q3)/(mo*co)#variação instântanea da temperatura do líquido
    dTrdt=(Q1-Q2)/(mr*cr)#variação instântanea da temperatura do recepiente
    dxdt=[dTodt,dTrdt]
    
    return dxdt
    
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

To=Ti
Tr=Ti
listinicial=[To,Tr]#Definindo os parâmetros de temperatura iniciais (0ºC ou 273.15k)

tempo1=np.arange(0,22800,120)#lista de tempo utilziada, de minutos para segundos
mododeint=odeint(modelo_caneca4,listinicial,tempo1)#função odeint utilizada
a1=[x-273.15 for x in mododeint[:,0]]#conversão de celsius para Kelvin
b1=[x-273.15 for x in mododeint[:,1]]#conversão de celsius para Kelvin
c1=[n/60 for n in tempo1]#conversão de segundos para minutos

ax=plt.axes()
plt.plot(c1,a1,'k')
ax.set_facecolor('xkcd:lightblue')
plt.title("Modelo com o melhor combo")
plt.xlabel('Tempo(min)')
plt.ylabel('Temperatura-Celsius')
plt.grid(True)
plt.show()

print('Assim, o tempo nesse combo para atingir a temperatura ambiente, com todos seus parâmetros ajustados, seria de {} minutos'.format(TempoAte23graus(a1,c1)))
```
